```python
import pandas as pd

tags = pd.read_csv('/content/lamoda_reviews_sampled_with_pull_tags.csv')
tags = tags[tags['tags'].notnull()]
```

```python
tags
```

```python
clothes = []
shoes = []
bags = []
beauty_accs = []
home_accs = []
toys = []
accs = []
jewellery = []

cats = ['Clothes', 'Shoes', 'Bags', 'Beauty_Accs', 'Home_Accs', 'Toys',
       'Accs', 'Jewellery']

for _, row in tags.iterrows():
    tag_list = [tag.strip() for tag in str(row['tags']).split(';') if tag.strip()]

    # Добавляем в соответствующий список в зависимости от категории
    if row['good_type'] == 'Clothes':
        clothes.extend(tag_list)
    elif row['good_type'] == 'Shoes':
        shoes.extend(tag_list)
    elif row['good_type'] == 'Bags':
        bags.extend(tag_list)
    elif row['good_type'] == 'Beauty_Accs':
        beauty_accs.extend(tag_list)
    elif row['good_type'] == 'Home_Accs':
        home_accs.extend(tag_list)
    elif row['good_type'] == 'Toys':
        toys.extend(tag_list)
    elif row['good_type'] == 'Accs':
        accs.extend(tag_list)
    elif row['good_type'] == 'Jewellery':
        jewellery.extend(tag_list)
```

```python
# Проверка с длиной в начале
print(len(set(clothes)), 'clothes:', set(clothes))
print(len(set(shoes)), 'shoes:', set(shoes))
print(len(set(bags)), 'bags:', set(bags))
print(len(set(beauty_accs)), 'beauty_accs:', set(beauty_accs))
print(len(set(home_accs)), 'home_accs:', set(home_accs))
print(len(set(toys)), 'toys:', set(toys))
print(len(set(accs)), 'accs:', set(accs))
print(len(set(jewellery)), 'jewellery:', set(jewellery))
```
