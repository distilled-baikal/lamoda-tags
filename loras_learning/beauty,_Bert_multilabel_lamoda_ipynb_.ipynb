```python
import torch
from transformers import AutoTokenizer, pipeline
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
from peft import get_peft_model, LoraConfig, TaskType
from urllib.request import urlopen
import io
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
#from datasets import load_dataset
from transformers import (
    DataCollatorWithPadding,
    TrainingArguments,
    Trainer,
)
import numpy as np
```

```python
# PyTorch + transformers
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

MODEL = "deepvk/user-bge-m3"
NUM_LABELS = num_cats

tokenizer = AutoTokenizer.from_pretrained(MODEL)
model_b = AutoModelForSequenceClassification.from_pretrained(MODEL,
                                                          num_labels=NUM_LABELS,
                                                          problem_type="multi_label_classification")
```

```python
#Lora настроить

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["query", "value"], #, "dense"], "key"
    lora_dropout=0.05,
    task_type=TaskType.SEQ_CLS
)

model = get_peft_model(model_b, lora_config)
```

```python
import torch, torch.nn.functional as F

# Инференс пример
texts = ["Это позитивный отзыв."]

enc = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

with torch.no_grad():
    out = model(**enc)

logits = out.logits
#print(logits)

probs = torch.sigmoid(logits)
preds = (probs > 0.5).int()

#print("logits:", logits)
print("probs:", probs)
print("preds:", preds)
```

```python
# 4) Токенизация

from torch.utils.data import Dataset, DataLoader
import torch

class SimpleTextDataset(Dataset):
    def __init__(self, data, tokenizer, max_length=512):
        """
        data: список кортежей (id, text, label)
        tokenizer: HuggingFace tokenizer
        """
        self.data = data
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        _id, text, label = self.data[idx]
        encoded = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',  # сразу паддинг до max_length
            max_length=self.max_length,
            return_tensors='pt'
        )
        # encoded возвращает тензоры с размерностью (1, seq_len), убираем лишний батч-изм
        item = {key: val.squeeze(0) for key, val in encoded.items()}
        item['labels'] = torch.tensor(label, dtype=torch.long)
        return item

'''
Пример датасета

dataset = [(0, 'Вкусный', [0,1,1,0,1]),
           (1, 'Плохой', [0,1,1,0,1]),
           (2, 'Хороший', [0,1,1,0,1])]
'''
```

```python
from tqdm import tqdm
import numpy as np

# Собираем в подходящий формат для SimpleTextDataset
#sub = np.random.choice(ds, 11000)
sub_filt = []

for i in tqdm(range(len(sub))):
    text = str(sub.iloc[i][0])

    label = list(sub.iloc[i])[2:]

    sub_filt.append((i, text, label))


sub_filt[0]
```

```python
# Проверка токенов

lens = []

for i in tqdm(range(len(sub_filt))):
    _, text, _ = sub_filt[i]

    lens.append(len(tokenizer(text)['input_ids']))

print(max(lens))


# Сколько больше, чем 768 токенов

res = 0

for len1 in lens:
    if len1 > 512: res += 1

print(res)
```

```python
import random

random.shuffle(sub_filt)

to_train = 1600

train_ds = sub_filt[:to_train]
val_ds   = sub_filt[to_train:]

train = SimpleTextDataset(train_ds, tokenizer)
val = SimpleTextDataset(val_ds, tokenizer)

tokenizer.decode(train[0]["input_ids"])
```

```python
import torch
from transformers import Trainer

name_dir = "./lora_bert_output_beauty"

class MultiLabelTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits
        loss_fct = torch.nn.BCEWithLogitsLoss()
        loss = loss_fct(logits, labels.float())
        return (loss, outputs) if return_outputs else loss

data_collator = DataCollatorWithPadding(tokenizer)

# Метрики
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

def compute_metrics(eval_pred):
    import numpy as np
    from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
    import torch

    logits, labels = eval_pred
    probs = torch.sigmoid(torch.tensor(logits)).numpy()   # превращаем в вероятности
    preds = (probs > 0.5).astype(int)                     # бинаризация

    return {
        "accuracy": accuracy_score(labels, preds),
        "f1": f1_score(labels, preds, average="micro"),   # micro лучше для multilabel
        "precision": precision_score(labels, preds, average="micro"),
        "recall": recall_score(labels, preds, average="micro"),
    }

training_args = TrainingArguments(
    output_dir=name_dir,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,

    eval_strategy="steps",    # или "epoch"
    eval_steps=50,            # если eval_strategy="steps"

    logging_strategy="steps",
    logging_steps=10,

    num_train_epochs=1,
    learning_rate=3e-4,
    weight_decay=0.0,
    fp16=True,

    load_best_model_at_end=True,
    metric_for_best_model="f1",
    greater_is_better=True,

    eval_on_start=True,       # запуск eval перед обучением
)

# 8) Trainer
trainer = MultiLabelTrainer(
    model=model,
    args=training_args,
    train_dataset=train,
    eval_dataset=val,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
```

```python
#Обучение, токен ae70255a743b6ad1c95f76562510685dd7b79a5d

trainer.train()

metrics = trainer.evaluate()
print(metrics)
```

```python
lora_path = name_dir

from transformers import AutoModelForSequenceClassification
from peft import PeftModel
from transformers import AutoTokenizer

base_model_name = "deepvk/user-bge-m3"  # тот же, что при обучении

base_model = AutoModelForSequenceClassification.from_pretrained(
    base_model_name,
    num_labels=num_cats   # ОБЯЗАТЕЛЬНО то же число классов
)

model_lora = PeftModel.from_pretrained(
    base_model,
    lora_path
)

tokenizer_lora = AutoTokenizer.from_pretrained(base_model_name)
```

```python
text = "Очень мягко очищает, хороший запах"

inputs = tokenizer_lora(
    text,
    return_tensors="pt",
    truncation=True,
    padding=True
)

with torch.no_grad():
    logits = model_lora(**inputs).logits

probs = torch.sigmoid(logits)[0]   # (23,)

threshold = 0.5

predicted_tags = [
    bag_tags[i] for i, p in enumerate(probs)
    if p >= threshold
]

print(predicted_tags)
```
